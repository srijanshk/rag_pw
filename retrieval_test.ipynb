```python
from FlagEmbedding import BGEM3FlagModel
from BGERetriever_v2 import BGERetriever
```

```python
import os
os.environ.setdefault("CUDA_VISIBLE_DEVICES", "3,4")
```

```python
faiss_index = "/local00/student/shakya/openmath_bge-m3_hnsw_index"
faiss_meta = "/local00/student/shakya/openmath_bge-m3_metadata.jsonl"

emb_model = BGEM3FlagModel("BAAI/bge-m3", use_fp16=True)
openmath_retriever = BGERetriever(
            embedding_model=emb_model,
            index_path=faiss_index,                    
            metadata_path=faiss_meta,            
)
```

```python
faiss_index = "/local00/student/shakya/mathpile_hnsw.index"
faiss_meta = "/local00/student/shakya/mathpile_meta.jsonl"

emb_model = BGEM3FlagModel("BAAI/bge-m3", use_fp16=True)
mathpile_retriever = BGERetriever(
            embedding_model=emb_model,
            index_path=faiss_index,                    
            metadata_path=faiss_meta,            
)
```

```python
query_norm = "Chinese Remainder Theorem"
k_dense = 100
k_final = 5
```

```python
mathpile_docs = mathpile_retriever.search_and_rerank(
				query_norm,
				top_k=k_dense,
				top_k_final=k_final,
		)

for doc in mathpile_docs:
		print(doc)
		print("----")
```

```python
openmath_docs = openmath_retriever.search_and_rerank(
				query_norm,
				top_k=k_dense,
				top_k_final=k_final,
		)

for doc in openmath_docs:
		print(doc)
		print("----")
```
